```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data
Y = iris.target

plt.figure(1, figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.figure(2, figsize=(8, 6))
plt.scatter(X[:, 2], X[:, 3], c=Y, edgecolor='k')
plt.xlabel('Petal length')
plt.ylabel('Petal width')

fig = plt.figure(3, figsize=(8, 6))
axis = Axes3D(fig, elev=-150, azim=20)
X_reduced = decomposition.PCA(n_components=3).fit_transform(iris.data)
axis.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=Y, edgecolor='k', s=40)
axis.set_title("3D visualization")
axis.set_xlabel("X")
axis.set_ylabel("Y")
axis.set_zlabel("Z")

plt.show()
```
